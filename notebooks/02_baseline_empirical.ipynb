# 02 — Empirical 20th-percentile baseline
Floor score. Every later model must beat this.

```python
import sys; sys.path.insert(0, str(__import__('pathlib').Path.cwd().parent))
import warnings; warnings.filterwarnings('ignore')
import pandas as pd
from src.data import load_or_build
from src.models.empirical import EmpiricalQuantileForecaster
from src.validation import DEFAULT_FOLDS, build_query_for_fold, evaluate
```

```python
ds = load_or_build()
rm_ids = sorted(ds.daily['rm_id'].unique().tolist())
for fold in DEFAULT_FOLDS:
    fc = EmpiricalQuantileForecaster(tau=0.2, min_year=2020).fit(ds.daily, history_end=fold.train_end + pd.Timedelta(days=1))
    preds = fc.predict(build_query_for_fold(fold, rm_ids))
    s = evaluate(preds, fold, ds.daily)
    print(f'{fold.name} pinball={s["mean_pinball"]:.0f}')
    print('  worst:', list(s['worst_rm_ids'].items())[:5])
```
